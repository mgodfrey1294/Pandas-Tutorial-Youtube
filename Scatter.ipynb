```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt

# Create DataFrame
my_df = pd.DataFrame(randn(500,4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df
```

```python
# Scatter Plot
my_df.plot(kind='scatter', x='Mon', y='Tues', c='Wed')
```

```python
# Colors: https://matplotlib.org/stable/tutorials/colors/colormaps.html
my_df.plot(kind='scatter', x='Mon', y='Tues', c='Wed', cmap='magma')
```

```python
# By Size
my_df.plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*200)
```

```python
# By Size easier to read with Alpha
my_df.plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*200, alpha=0.4)
```

```python
# Title Etc
my_df.plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*200, alpha=0.4, title='My Awesome Scatter Plot')
```
