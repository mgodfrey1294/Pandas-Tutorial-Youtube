```python
import pandas as pd
import numpy as np
# Import CSV File
my_df = pd.read_csv('https://raw.githubusercontent.com/flatplanet/Pandas-Youtube-Playlist/main/dog_data.csv')
my_df
```

```python
# Conditional Selection
# > < = >= <= etc
# Return Booleans
my_df == "BROWN"
```

```python
# Return Dataframe with data
my_df[my_df == "BROWN"]
```

```python
# Run them on a column
my_df[my_df == "BROWN"]["Color"]
```

```python
# Return Multiple Columns
my_df[my_df == "BROWN"][["Color", "Breed"]]
```
