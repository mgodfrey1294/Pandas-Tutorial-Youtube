```python
import pandas as pd
import numpy as np
# Import CSV File
my_df = pd.read_csv('https://raw.githubusercontent.com/flatplanet/Pandas-Youtube-Playlist/main/dog_data.csv')
my_df
```

```python
# Grab a specific column
my_df['DogName']
```

```python
# Count Values
my_df['DogName'].value_counts()
# Pass as DataFrame
pd.DataFrame(my_df['DogName'].value_counts()).head(50)
```

```python
# Grab Uniques
my_df['DogName'].unique()
```

```python
# Show as DataFrame
pd.DataFrame(my_df['DogName'].unique()).head(50)
```
