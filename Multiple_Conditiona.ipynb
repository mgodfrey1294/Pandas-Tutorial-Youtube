```python
import pandas as pd
import numpy as np
# Import CSV File
my_df = pd.read_csv('https://raw.githubusercontent.com/flatplanet/Pandas-Youtube-Playlist/main/dog_data.csv')
my_df
```

```python
# Multiple Conditional AND
my_df[(my_df["Color"] == "BROWN") & (my_df["Breed"] == "MIXED")]
```

```python
# Get Length
len(my_df[(my_df["Color"] == "BROWN") & (my_df["Breed"] == "MIXED")])
```

```python
# Multiple Conditional OR
my_df[(my_df["Color"] == "BROWN") | (my_df["Breed"] == "MIXED")]
```

```python
# Get Length
len(my_df[(my_df["Color"] == "BROWN") | (my_df["Breed"] == "MIXED")])
```

```python
# Return Specific Column
my_df[(my_df["Color"] == "BROWN") & (my_df["Breed"] == "MIXED")]["DogName"]
```
