```python
import pandas as pd
import numpy as np
# Import CSV File
my_df = pd.read_csv('https://raw.githubusercontent.com/flatplanet/Pandas-Youtube-Playlist/main/dog_data.csv')
# Create a smaller dataframe that's easier to work with
my_df = my_df.loc[:4]
my_df
```

```python
# Create New Column
my_df["Frame Header"] = ["Dog 1", "Dog 2", "Dog 3", "Dog 4", "Dog 5"]
my_df
```

```python
# Set Index
my_df.set_index("Frame Header", inplace=True)
```

```python
# Show DataFrame
my_df
```

```python
# Reset Index
my_df.reset_index(inplace=True)
```

```python
# Drop Column
my_df.drop("Frame Header", axis=1, inplace=True)
```

```python
# Show DataFrame
my_df
```
