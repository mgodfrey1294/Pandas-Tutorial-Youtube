```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt

# Create DataFrame
my_df = pd.DataFrame(randn(25,4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df
```

```python
# Area plot
my_df.abs().plot(kind='area', stacked=False)
```

```python
# Shading
my_df.abs().plot(kind='area', alpha=0.5)
```

```python
# Title
my_df.abs().plot(kind='area', alpha=0.5, title='My Awesome Area Plot!')
```

```python
# Legend
my_df.abs().plot(kind='area', alpha=0.5, title='My Awesome Area Plot!', legend=False)
```

```python
# Plot only one column
my_df['Mon'].abs().plot(kind='area', alpha=0.5, title='My Awesome Area Plot!', legend=False)
```

```python
# Plot 2 columns
my_df[['Mon', 'Thurs']].abs().plot(kind='area', alpha=0.5, title='My Awesome Area Plot!', legend=False)
```

```python
# Add Table
my_df.abs().plot(kind='area', alpha=0.5, title='My Awesome Area Plot!', legend=False, table=True)
```
