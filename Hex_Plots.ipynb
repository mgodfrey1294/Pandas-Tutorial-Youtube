```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt

# Create DataFrame
my_df = pd.DataFrame(randn(500,4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df
```

```python
# Hexplots
my_df.plot(kind='hexbin', x='Mon', y='Tues')
```

```python
# Hexplots Gridsize
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=10)
```

```python
# Hexplots Color
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=10, cmap='magma')
```

```python
# Other Method
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=10)
```

```python
# Title
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=10, title='My Awesome Hex Plot')
```

```python
# Grid
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=10, title='My Awesome Hex Plot', grid=True)
```
