```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt

# Create DataFrame
my_df = pd.DataFrame(randn(25,4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df
```

```python
# Bar Plot
my_df.abs().plot(kind='bar')
```

```python
# Bar Plot Stacked
my_df.abs().plot(kind='bar', stacked=True)
```

```python
# Shading
my_df.abs().plot(kind='bar', stacked=True, alpha=0.4)
```

```python
# Title
my_df.abs().plot(kind='bar', stacked=True, title='My Awesome Bar Chart')
```

```python
# Legend
my_df.abs().plot(kind='bar', stacked=True, title='My Awesome Bar Chart', legend=False)
```

```python
# Second Way
my_df.abs().plot.bar()
```

```python
# Second Way Stacked
my_df.abs().plot.bar(stacked=True)
```

```python
# Second Way Shading
my_df.abs().plot.bar(stacked=True, alpha=0.4)
```

```python
# Second Way Title
my_df.abs().plot.bar(stacked=True, title='My Bodacious Bar Chart')
```

```python
# Second Way Legend
my_df.abs().plot.bar(stacked=True, legend=False)
```

```python
# Grid Lines
my_df.abs().plot.bar(stacked=True, grid=True)
```
