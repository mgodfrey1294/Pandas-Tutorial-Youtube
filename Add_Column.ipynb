```python
import pandas as pd
import numpy as np
# import csv file
my_df = pd.read_csv('https://raw.githubusercontent.com/flatplanet/Pandas-Youtube-Playlist/main/dog_data.csv')
my_df
```

```python
# Add Column From List
my_df2 = my_df.iloc[:5,]
my_df2
gender = ["Male", "Female", "Male", "Male", "Female"]
my_df2["Gender"] = gender
my_df2
```

```python
# Add Default Values
my_df["Alive/Dead"] = [True]*len(my_df)
my_df
```

```python
# Add NaN Value
my_df["Show Dog"] = [np.nan]*len(my_df)
my_df
```

```python
# Add column with .insert() - Allows Position
my_df.insert(1, "Adopted", [True]*len(my_df), True)
my_df
```

```python
# Add column with .assign() - Creates new df
my_df2 = my_df.assign(Horse=[False]*len(my_df))
my_df2
```
